# 3. Data Preparation

## 3.1 Обработка выбросов
На гистограммах и парных распределениях видны несколько выбросов. Так как их количество невелико, то предлагается исключить их из  датасета.

```python
vine[(vine['citric acid'] < 0.75) & 
     (vine['residual sugar'] < 20) & 
     (vine['chlorides'] < 0.4) & 
     (vine['free sulfur dioxide'] < 125) & 
     (vine['total sulfur dioxide'] < 300) &
     ~((vine['type'] == 'red') & ((vine['total sulfur dioxide'] > 220))) &
     ~((vine['type'] == 'red') & ((vine['residual sugar'] > 10))) &
     (vine['density'] < 1.05) &
     (vine['sulphates'] <  1.5) &
     (vine['alcohol'] !=  8) &
     (vine['alcohol'] <  14)
    ].shape
```

Такая фильтрация уменьшит датасет на 2.5 процента. Автор предполагает, что это приведет к улучшению категоризации оставшихся после фильтрации объектов

```python
vine_filtererd = vine[(vine['citric acid'] < 0.75) & (vine['residual sugar'] < 20) &
                      (vine['chlorides'] < 0.4) & 
                      (vine['free sulfur dioxide'] < 125) & 
                      (vine['total sulfur dioxide'] < 300) &
                      ~((vine['type'] == 'red') & ((vine['total sulfur dioxide'] > 220))) &
                      ~((vine['type'] == 'red') & ((vine['residual sugar'] > 10))) &
                      (vine['density'] < 1.05) &
                      (vine['sulphates'] <  1.5) &
                      (vine['alcohol'] !=  8) &
                      (vine['alcohol'] <  14)
                     ]
vine_filtererd.reset_index(drop=True,inplace = True)
```

 Кажется разумным взглянуть на графики взаимных распередений и корреляций после фильтрации

## 3.2 Категориальные переменные должны быть преобразованы в числовые переменные

Переменная Type рассматривается как категориальная переменная. Некоторые из  алгоритмов могут обрабатывать только числовые значения, поэтому нам нужно создать новую (фиктивную) переменную для каждого уникального значения категориальных переменных (OneHotEncoding)

```python
vine_type = pd.get_dummies(vine_filtererd.type , drop_first = True).reset_index(drop=True)
vine_type.shape
```

## 3.3 Feature Engineering &ndash; добавляем новые признаки


### 3.3.1 Создание новых признаков из комбинации старых
Кораллелограмма отчетливо показывает наличие взаимных зависимостей различных характеристик.  Использование данных в таком виде для линейных моделей может привести к неустойчивому решению. Борьбой с эти может быть как введение регуляризации, так и снятие взаимных зависимостей. Предлагается создание новых факторов на базе старых характеристик в виде комбинаций произведений. При  это использованные для создания новых факторов  характеристики изымаются из датасета. Таким образом предлагается уменьшить взаимную корреляцию признаков, используемых в датасете при создании модели. Среди получившихся факторов также проводит проверка дублирование: если комбинация признаков в данном факторе не уникальна ( нет хотя бы одного уникального признака для данного фактора по сравнению со всеми признаками остальных факторов) - данный признак удаляется.

```python
factors_df = pd.DataFrame()
used_col = []

vine_filtererd[list_columns[:-1]]
col_pref = 'Factor_'
factor_dict = {}
factor_n = 1
list_factors = []
used_col_singl = []
for col1 in list_columns[:-1]:
    new_factor = True
    fact = []
    for col2 in list_columns[:-1]:
        corr_cols = vine_filtererd[col1].corr(vine_filtererd[col2])        
        if ~((col2+col1) in used_col) & (col1 != col2) & (abs(corr_cols) >0.4):
            if new_factor:
                factor_name = col_pref + str(factor_n)
                factors_df[factor_name] = vine_filtererd[col1]
                factor_n += 1
                new_factor = False
                list_factors.append(factor_name)
                fact.append(col1)
                used_col_singl.append(col1)
            factors_df[factor_name] = factors_df[factor_name] * vine_filtererd[col2] 
            used_col.append(col1+col2)
            fact.append(col2)
            used_col_singl.append(col2)
    if len(fact) != 0:
        factor_dict.update( {factor_name : fact} )
    
drop_cols = []
for key_factor in list_factors:
    another_factors = [element for element in list_factors  if element not in [key_factor]]
    all_used =list(set(list(itertools.chain(*[factor_dict[x] for x in another_factors]))))
    important = [element for element in factor_dict[key_factor]  if element not in all_used]
    if len(important) == 0:
        drop_cols.append(key_factor)
```

 Состав факторов по парметрам

```python
factor_dict
```

```python
factors_df.drop(drop_cols, axis=1, inplace = True)
factors_df.reset_index(drop=True,inplace = True)
used_columns = list(set(used_col_singl))
```

```python
plt.figure(figsize=(10,10), dpi= 80)
ax = sns.heatmap(factors_df.corr(), 
                 xticklabels=factors_df.corr().columns, 
                 yticklabels=factors_df.corr().columns, 
                 cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
load_ipython_extension(plt)
```

## 3.4 Сборка финальных датасетов для моделирования

### 3.4.1 Variable selection
Выбираем признаки для формирования итогового датасета. 

```python
used_columns.append('type')
used_columns.append('quality')
full_X = factors_df.join(vine_filtererd.drop(columns=used_columns)).join(vine_type )
full_X.head()
```

```python
used_columns.append('type')
used_columns.append('quality')
full_X = factors_df.join(vine_filtererd.drop(columns=used_columns), how = 'inner').join(vine_type, how = 'inner' )
full_X.head()
```

```python

plt.figure(figsize=(10,10), dpi= 80)
ax = sns.heatmap(full_X.corr(), 
                 xticklabels=full_X.corr().columns, 
                 yticklabels=full_X.corr().columns, 
                 cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
load_ipython_extension(plt)
```

### 3.4.2 Создание датасетов

Отделяем данные для обучения и для проверки

```python

X = full_X
y = vine_filtererd.quality
print(X.shape, y.shape)
```

```python

X = full_X
sc = StandardScaler()
X = sc.fit_transform(X)
y = vine_filtererd.quality
train_X , test_X , train_y , test_y = train_test_split( X , y , train_size = .7, random_state =2020)
```

```python


print (full_X.shape , train_X.shape ,  train_y.shape , test_y.shape , test_X.shape)
```

# 4. Моделирование

## 4.1 Выбор модели
Был выбран вариант OneVsRestClassifier с SVC классификатором с ядром rbf и балансировкой классов

```python

model = OneVsRestClassifier(SVC(kernel="rbf", class_weight = 'balanced'))
```

## 4.2 Обучение модели

```python
model.fit( train_X , train_y )
```

# 5. Оценка

Оценить точность модели возможно, используя набор для тестирования, где известен фактический результат. Этот набор данных не использовался для обучения, поэтому он абсолютно новый для модели.

Если сравнить точность предсказания на тестовых и тренировочных данных, можно  оценить качество работы модели с точки зрения переобучения. Если разница между ними значительна, это свидетельствует о переобучении.


```python
# Score the model
print (model.score( train_X , train_y ) , model.score( test_X , test_y ))
```

Также для установления качества работы модели можно воспользоваться  подробным отчетом по прогнозированию классификации образцов по каждому классу отдельно.

```python

print(classification_report( test_y, model.predict(test_X)))
```

```python
unique_y = y.unique()
unique_y.sort()
unique_y
class_name= [f'Class_{x}' for x in unique_y]
pred_class_name = [f'Pred_{x}' for x in unique_y]

confusion_df = pd.DataFrame(columns = class_name, index = pred_class_name, 
                            data = confusion_matrix( test_y, model.predict(test_X)))
confusion_df
```

Более подробный отчёт показывает, что качество работы модели падает пропорционально количеству образцов класса в тестовой выборке. Так же заметно смещение матожидания каждого класса к матожиданю всей выборки
